```python
import random

# Constants
NUM_ROUNDS = 5  # Number of rounds

# Welcome message
print("🎮 Welcome to the High👋↗️ and Low↙️ game!!!")
print("------------------------------------------")

# Initialize score
score = 0

# Game loop
for round_num in range(1, NUM_ROUNDS + 1):
    print(f"\n🔥 Round {round_num} 🔥")

    # Step 1: Generate numbers
    user_num = random.randint(1, 100)
    computer_num = random.randint(1, 100)

    # Show the user their number
    print(f"🔢 Your number is: {user_num}")

    # Step 2: Get user input (with input validation)
    while True:
        guess = input("Do you think your number is HIGHER or LOWER than the computer's? (Type: higher/lower) ➡ ").strip().lower()

        if guess == "higher" or guess == "lower":
            break  # Exit loop if input is correct

        print("❌ Invalid input! Please enter either 'higher' or 'lower'.")

    # Step 3: Game logic
    if (guess == "higher" and user_num > computer_num) or (guess == "lower" and user_num < computer_num):
        print(f"✅ Correct! The computer's number was {computer_num}.")
        score += 1  # Increase score if correct
    else:
        print(f"❌ Wrong! The computer's number was {computer_num}.")

    print(f"📊 Your score is now: {score}\n")

# Step 4: Final message based on performance
print("🎮 Thanks for playing!")

if score == NUM_ROUNDS:
    print("🏆 Wow! You played perfectly! 🎉")
elif score >= NUM_ROUNDS // 2:
    print("👏 Good job, you played really well! 👍")
else:
    print("😢 Better luck next time! 👋")
```
